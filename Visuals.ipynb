```python
import pandas as pd
```

```python
import os
os.listdir()
```

```python
df = pd.read_excel('County Health Data 2021.xlsx')
```

```python
melted = pd.melt(df, ['Unnamed: 0', 'Unnamed: 1', 'Unnamed: 2','Unnamed: 3','Unnamed: 4', 'Data year', 'Data Notes'])
```

```python
melted = melted.rename({'Unnamed: 0':'feature', 'Unnamed: 1': 'feature_longer', 'Unnamed: 2':'source', 'Unnamed: 3':'link','Unnamed: 4':'etc',
                       'Data year':'data_year', 'Data Notes':'data_notes', 'variable':'county'}, axis = 1)
```

```python
melted
```

```python
data_options = ['Life expectancy', 'Food Insecurity', 'Unemployment']
```

```python
selected_data = melted[melted['feature'] == data_options[0]]
```

```python
county_options = melted['county'].unique()
```

```python
single_data = selected_data[selected_data['county'] == county_options[0]]
```

```python
selected_data
```

```python
import altair as alt
import numpy as np
```

```python
alt.Chart(selected_data).mark_bar().encode(
    alt.X("value:Q", bin=True),
    y='count()',
).properties(title = 'Distribution of {}'.format(data_options[0].lower()))
```

```python
colors = ['red' if x == 'Alamance' else 'blue' for x in selected_data.sort_values(by = 'value', ascending = False).county]
```

```python
def apply_color(county):
    if county == 'Alamance':
        return 'Selected County: {}'.format('Alamance')
    else:
        return 'All Others'
```

```python
selected_data['color'] = selected_data['county'].map(apply_color)
```

```python
alt.Chart(selected_data.sort_values(by = 'value', ascending = False)).mark_bar().encode(
    alt.Y("county", sort='-x', axis=alt.Axis(labels = False)),
    alt.X("value:Q"),
color = 'color').properties(
    width=400,
    height=600
)
```

```python
data_points = {'Average':np.mean(selected_data['value']), county_options[0]:single_data.reset_index().loc[:,'value'][0]}
```

```python
data = {'county':['Average', county_options[0]], 'values':[np.mean(selected_data['value']), single_data.reset_index().loc[:,'value'][0]]}
data_points = pd.DataFrame.from_dict(data)
```

```python
alt.Chart(data_points).mark_bar().encode(
    y='county',
    x=alt.X('values')
).properties(height = 100)
```

```python
import altair as alt
from vega_datasets import data
counties = alt.topo_feature(data.us_10m.url, 'counties')
```

```python
alt.Chart(counties).mark_geoshape().encode(
    color='value:Q',
).transform_lookup(
    lookup='id',
    from_=alt.LookupData(selected_data, 'id', ['value'])
).project(
    type='albersUsa'
).properties(
    width=500,
    height=300
)
```

```python
import altair as alt
from vega_datasets import data

counties = alt.topo_feature(data.us_10m.url, 'counties')
source = data.unemployment.url

alt.Chart(counties).mark_geoshape().encode(
    color='rate:Q'
).transform_lookup(
    lookup='id',
    from_=alt.LookupData(source, 'id', ['rate'])
).project(
    type='albersUsa'
).properties(
    width=500,
    height=300
)
```
